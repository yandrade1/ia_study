# Semana 2 - Dia 2: Distribuições de Probabilidade
Exemplos práticos com distribuições discretas e contínuas usando NumPy e SciPy.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, uniform, norm, expon

plt.style.use('seaborn-v0_8')
```

## 1. Distribuições Discretas

```python
# Bernoulli
data = bernoulli.rvs(p=0.5, size=1000)
plt.hist(data, bins=2, rwidth=0.8)
plt.title("Bernoulli (p=0.5)")
plt.show()

# Binomial
data = binom.rvs(n=10, p=0.5, size=1000)
plt.hist(data, bins=11, rwidth=0.8)
plt.title("Binomial (n=10, p=0.5)")
plt.show()

# Poisson
data = poisson.rvs(mu=3, size=1000)
plt.hist(data, bins=15, rwidth=0.8)
plt.title("Poisson (λ=3)")
plt.show()
```

## 2. Distribuições Contínuas

```python
# Uniforme
data = uniform.rvs(loc=0, scale=10, size=1000)
plt.hist(data, bins=20, rwidth=0.8)
plt.title("Uniforme (0,10)")
plt.show()

# Normal
data = norm.rvs(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, rwidth=0.8)
plt.title("Normal (μ=0, σ=1)")
plt.show()

# Exponencial
data = expon.rvs(scale=1, size=1000)
plt.hist(data, bins=30, rwidth=0.8)
plt.title("Exponencial (λ=1)")
plt.show()
```

## O que esse bloco faz:

Gera 10.000 amostras de cada distribuição.

Calcula média e variância teórica (da fórmula).

Calcula média e variância observada (dos dados simulados).

Imprime comparando os dois valores lado a lado.

Você vai ver que quanto mais amostras usar, mais os valores observados vão “grudar” nos esperados → lei dos grandes números em ação.

```python
print("=== Comparação Média e Variância ===\n")

# Bernoulli
p = 0.5
data = bernoulli.rvs(p=p, size=10000)
print("Bernoulli(p=0.5)")
print("Média esperada:", p, " | Média observada:", np.mean(data))
print("Var esperada:", p*(1-p), " | Var observada:", np.var(data))
print("-"*40)

# Binomial
n, p = 10, 0.5
data = binom.rvs(n=n, p=p, size=10000)
print("Binomial(n=10, p=0.5)")
print("Média esperada:", n*p, " | Média observada:", np.mean(data))
print("Var esperada:", n*p*(1-p), " | Var observada:", np.var(data))
print("-"*40)

# Poisson
lam = 3
data = poisson.rvs(mu=lam, size=10000)
print("Poisson(λ=3)")
print("Média esperada:", lam, " | Média observada:", np.mean(data))
print("Var esperada:", lam, " | Var observada:", np.var(data))
print("-"*40)

# Uniforme
a, b = 0, 10
data = uniform.rvs(loc=a, scale=b-a, size=10000)
print("Uniforme(0,10)")
print("Média esperada:", (a+b)/2, " | Média observada:", np.mean(data))
print("Var esperada:", (b-a)**2/12, " | Var observada:", np.var(data))
print("-"*40)

# Normal
mu, sigma = 0, 1
data = norm.rvs(loc=mu, scale=sigma, size=10000)
print("Normal(μ=0, σ²=1)")
print("Média esperada:", mu, " | Média observada:", np.mean(data))
print("Var esperada:", sigma**2, " | Var observada:", np.var(data))
print("-"*40)

# Exponencial
lam = 1
data = expon.rvs(scale=1/lam, size=10000)
print("Exponencial(λ=1)")
print("Média esperada:", 1/lam, " | Média observada:", np.mean(data))
print("Var esperada:", 1/(lam**2), " | Var observada:", np.var(data))
print("-"*40)
```
